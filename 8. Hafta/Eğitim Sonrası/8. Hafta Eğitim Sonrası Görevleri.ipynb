# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
girisSayisi = int(input("Kaç giriş yapacaksınız? :"))
isimler = []
soyisimler = []
yaslar = []
boylar = []
kilolar = []

for i in range(girisSayisi):
    isimler.append(input("İsim: "))
    soyisimler.append(input("Soyisim: "))
    yaslar.append(int(input("Yaş: ")))
    boylar.append(int(input("Boy (cm): ")))
    kilolar.append(int(input("Kilo (kg): ")))

file = open("VkiListesi.txt", "w", encoding="utf-8")

class Zayif:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Zayıf Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class Normal:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Normal Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class FazlaKilolu:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Fazla Kilolu Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class Obez:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Obez Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class MorbidObez:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Morbid Obez Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")


for i in range(girisSayisi):
    boy_metre = boylar[i] / 100
    vki = kilolar[i] / (boy_metre ** 2)

    if vki < 18.5:
        kisi = Zayif(isimler[i], soyisimler[i], yaslar[i], vki)
    elif 18.5 <= vki < 24.9:
        kisi = Normal(isimler[i], soyisimler[i], yaslar[i], vki)
    elif 25 <= vki < 29.9:
        kisi = FazlaKilolu(isimler[i], soyisimler[i], yaslar[i], vki)
    elif 30 <= vki < 39.9:
        kisi = Obez(isimler[i], soyisimler[i], yaslar[i], vki)
    else:
        kisi = MorbidObez(isimler[i], soyisimler[i], yaslar[i], vki)

    kisi.yazdir(file)

file.close()
```
