# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
girisSayisi = int(input("Kaç giriş yapacaksınız? :"))
isimler = []
soyisimler = []
yaslar = []
boylar = []
kilolar = []

for i in range(girisSayisi):
    isimler.append(input("İsim: "))
    soyisimler.append(input("Soyisim: "))
    yaslar.append(int(input("Yaş: ")))
    boylar.append(int(input("Boy (cm): ")))
    kilolar.append(int(input("Kilo (kg): ")))

file = open("VkiListesi.txt", "w", encoding="utf-8")

class Zayif:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Zayıf Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class Normal:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Normal Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class FazlaKilolu:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Fazla Kilolu Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class Obez:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Obez Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")

class MorbidObez:
    def __init__(self, isim, soyisim, yas, vki):
        self.isim = isim
        self.soyisim = soyisim
        self.yas = yas
        self.vki = vki

    def yazdir(self, file):
        file.write(f"""Morbid Obez Kişiler:
İsim: {self.isim} 
Soyisim: {self.soyisim}
Yas: {self.yas}
\n""")


for i in range(girisSayisi):
    boy_metre = boylar[i] / 100
    vki = kilolar[i] / (boy_metre ** 2)

    if vki < 18.5:
        kisi = Zayif(isimler[i], soyisimler[i], yaslar[i], vki)
    elif 18.5 <= vki < 24.9:
        kisi = Normal(isimler[i], soyisimler[i], yaslar[i], vki)
    elif 25 <= vki < 29.9:
        kisi = FazlaKilolu(isimler[i], soyisimler[i], yaslar[i], vki)
    elif 30 <= vki < 39.9:
        kisi = Obez(isimler[i], soyisimler[i], yaslar[i], vki)
    else:
        kisi = MorbidObez(isimler[i], soyisimler[i], yaslar[i], vki)

    kisi.yazdir(file)

file.close()
```

**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time

satinAlinanlar = []

def satinAlma(urun_fiyat):
    try:
        verilen_para = float(input("Ödeyeceğiniz miktar: "))  
        if urun_fiyat > verilen_para:
            print("Yetersiz Ödeme!")
        else:
            print(f"Afiyet olsun! Para üstünüz: {verilen_para - urun_fiyat} TL")
            
            satinAlinanlar.append(urun_fiyat)
    except ValueError:
        print("Lütfen geçerli bir sayı girin.")  


def satinAlmaGecmisi():
    if len(satinAlinanlar) == 0:
        print("Henüz bir satın alma işlemi gerçekleştirmediniz.")
    else:
        print("Satın Alma Geçmişi:")
        for index, fiyat in enumerate(satinAlinanlar, start=1):
            print(f"{index}. Smoothie - Fiyat: {fiyat} TL")


def menüGöster():
    print(""" Menü :
           1. Detoks Smoothie:               170 TL
           2. Protein Smoothie:              200 TL   
           3. Kırmızı Meyveler Smoothie:     200 TL
           """)


def smoothieHazirla():
    print("Smoothie hazırlamak sabit fiyattır. 210 TL")
    meyvesecim = input("""İstediğiniz Meyveler: (birden çok seçebilirsiniz):
                        1. Çilek
                        2. Muz
                        3. Ananas
                        4. Elma""")
    
    sivisecim = input("""Sıvı seçiminiz:
                       1. Yulaf sütü
                       2. Laktozsuz Süt
                       3. Süt
                       4. Yoğurt""")
    
    tatlandirici = input("""Tatlandırıcı seçimi:
                          1. Bal
                          2. Akçaağaç Şurubu
                          3. Şeker
                          4. Tatlandırıcı yok""")
    
    ekstra = input("""Ekstra malzemeler:
                    1. Chia Tohumu
                    2. Yulaf
                    3. Protein Tozu
                    4. Ekstra Malzeme İstemiyorum""")
    
    print("Smoothiniz Hazırlanıyor...")
    time.sleep(2)
    print("Smoothiniz Hazır.. Ödeme Aşamasına Geçebiliriz.")
    
    
    return 210


def main():
    print("""Hoşgeldiniz... 
             İsterseniz menüden hazır smoothilere bakabilir, isterseniz kendi smoothinizi hazırlayabilirsiniz.
             Ne yapmak istersiniz ?
             1. Menüye bak.
             2. Kendi smoothinizi hazırla.
             3. Satın alma geçmişini görüntüle.
             4. Çıkış için 'q' tuşuna basınız.""")
    
    while True:
        secim = input()

        if secim == "1":
            menüGöster()
        elif secim == "2":
            smoothie_fiyat = smoothieHazirla()  
            satinAlma(smoothie_fiyat)  
        elif secim == "3":
            satinAlmaGecmisi()  
        elif secim == "q":
            break  

if __name__ == "__main__":
    main()
    


    
    
    
```
